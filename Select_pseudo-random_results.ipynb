```python
import pandas as pd
import csv
import numpy as np
import os
```

```python
path = "./AMT_results/"
```

```python
def csv_to_DF():
    all_video_pairs = pd.DataFrame()
    for csv_file in os.listdir(path):
        if csv_file.startswith("videopair_"):
            new_csv_df = pd.read_csv(path+csv_file)
            all_video_pairs = all_video_pairs.append(new_csv_df)
    return all_video_pairs
```

```python
def take_100_pseudorand():
    DF100 = all_video_pairs.head(1)
    display(DF100)
    nonmatches = 1
    while len(DF100) < 100:
        for i, sample  in all_video_pairs.iterrows():
            if sample.isna().any():
                continue
            sample_traits = sample[['authentic','organized','comfortable','imaginative','interview']]
            DF100_traits = DF100[['authentic','organized','comfortable','imaginative','interview']]
            if (DF100_traits != sample_traits).sum(axis=1).min() >= nonmatches:
                DF100 = DF100.append(sample)
                display(pd.DataFrame(sample).transpose())
                if len(DF100) == 100:
                    break
        nonmatches = nonmatches - 1
        print nonmatches, len(DF100)
        
    return DF100
```

```python
csv_to_DF()
DF100 = take_100_pseudorand()
```
